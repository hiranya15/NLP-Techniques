```python
#1. Tokenization
import nltk #national language tool kit
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
text= "Natural language processing is interesting"
tokens= word_tokenize(text)
print(tokens)
```

```python
#sentence tokenization
from nltk.tokenize import sent_tokenize
text = "Python is easy. It is widely used in data science."

sentences = sent_tokenize(text)
print(sentences)
```

```python
#Stop word removal
#Stop words such as is,the,and, in are removed
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

words= word_tokenize(text)
filtered_words = [w for w in words if w.lower() not in stop_words]
print(filtered_words)
```

```python
#Stemming
#Stemming reduces words to their root form
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["running", "playing", "studies", "coding", "sleeping"]

for w in words:
  print(stemmer.stem(w))
```

```python
#Lemmatization
#It converts words to their dictionary form
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("writing"))
print(lemmatizer.lemmatize("studies"))
print(lemmatizer.lemmatize("best"))
print(lemmatizer.lemmatize("better"))
print(lemmatizer.lemmatize("eating"))
```

```python
#Part-of-speech(POS) Tagging
#POS tagging identifies grammatical roles in words

import nltk
nltk.download('average_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
words = word_tokenize("Python is a powerful programming language")
pos_tags =pos_tag(words)
print(pos_tags)
```

```python
#Word Frequency count
from nltk.probability import FreqDist

words = word_tokenize(text)

fd = FreqDist(words)

print(fd.most_common())
```
